exploring the Gapminder data

importing necessary libraries and the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
data_url = "https://raw.githubusercontent.com/risitadas/Machine-Learning-with-scikit-learn/main/gapminder_full-csv.csv"
gapminder = pd.read_csv(data_url)
gapminder.head()
```

plotting a simple graph from our data

```python
plt.plot(gapminder.gdp_cap, gapminder.life_exp)
plt.show()
```

further making a simpler scatter plot

```python
plt.scatter(gapminder.gdp_cap, gapminder.life_exp)

plt.xscale('log')
plt.show()
```

therfore, higher GDP usually corresponds to a higher life expectancy, there is a positive correlation

further creating a histogram of life expectency

```python
plt.hist(gapminder.life_exp)
plt.show()
```

with more bins for a better visibility

```python
plt.hist(gapminder.life_exp, bins=20)
plt.show()
```

furthur improving the scatter plot

```python
plt.scatter(gapminder.gdp_cap, gapminder.life_exp)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.show()
```

```python
plt.scatter(gapminder.gdp_cap, gapminder.life_exp)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.show()
```

storing population as a numpy array : np_pop

```python
np_pop = np.array(gapminder.population)
np_pop2 = np_pop*2
```

for better customization, using seaborn scatterplot

```python
sns.scatterplot(gapminder['gdp_cap'], gapminder['life_exp'], hue = gapminder['continent'], size = np_pop2, sizes=(20,400))
plt.grid(True)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.show()
```

increasing the graph size

```python
plt.figure(dpi=150)
```

storing population as a numpy array : np_pop

```python
np_pop = np.array(gapminder.population)
np_pop2 = np_pop*2
```

removing the legend

```python
sns.scatterplot(gapminder['gdp_cap'], gapminder['life_exp'], hue = gapminder['continent'], size=np_pop2, sizes=(20,400), legend = False)
plt.grid(True)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]', fontsize = 14)
plt.ylabel('Life Expectancy [in years]', fontsize = 14)
plt.title('World Development in 2007', fontsize = 20)
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.show()
```

adding description to the biggest countries

```python
plt.figure(dpi=150)
np_pop = np.array(gapminder.population)
np_pop2 = np_pop*2
sns.scatterplot(gapminder['gdp_cap'], gapminder['life_exp'], hue = gapminder['continent'], legend = False, 
                size=np_pop2, sizes=(20,500), alpha = 0.8)
plt.grid(True)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]', fontsize = 14)
plt.ylabel('Life Expectancy [in years]', fontsize = 14)
plt.title('World Development in 2007', fontsize = 20)
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
#A
plt.text(1550, 67, 'India')
plt.text(5650, 75, 'China')
plt.show()
```
